```python

np.random.seed(0)
errors = []
n_values = []
for i in range(10,100000,3):
    n = i
    x = np.random.uniform(-1, 1, n)
    y = np.random.uniform(-1, 1, n)
    d  = np.sqrt(x**2 + y**2)
    inside = d <= 1
    pi = 4 * np.sum(inside) / n
    error = (np.pi - pi)  
    errors.append(error)
    n_values.append(n)
plt.figure(figsize=(10, 6))
plt.plot(n_values, errors, marker='o', linestyle='-', color='blue', label='Error in π estimate')
plt.axhline(0, color='green', linestyle='--', label='True Mean (0)')
plt.xscale('log')
```

```python
np.random.seed(20)
N = 10000
S = [10, 50, 100, 1000 ]
for i in range (3):
    plt.figure()
    for n in S:
        means1 = []
        means2 = []
        means3 = []
        for a in range(N):
            samples1 =np.random.normal(0, 1, n)
            means1.append(np.mean(samples1))     
            samples2 = np.random.uniform(-1, 1, n)
            means2.append(np.mean(samples2))     
            samples3 = np.random.standard_cauchy(n)  
            means3.append(np.mean(samples3))     
        if(i==0):
            plt.hist(means1, bins=20, alpha=0.5, density=True)
            plt.title("Normal Distribution")
            print("mean normal for n =  ",n,": ",np.sum(means1)/N)
            print("mean uniform for n =   ",n,": ",np.sum(means2)/N)
            print("mean cauchy for n =   ",n,": ",np.sum(means3)/N)    
        elif(i==1):
            plt.title("Uniform Distribution")
            plt.hist(means2, bins=20, alpha=0.5, density=True)
        else:
            plt.title("Cauchy Distribution")
            plt.hist(means3, bins=20, alpha=0.5, density=True)    

    plt.show()

```
